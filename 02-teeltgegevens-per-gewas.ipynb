```python
import pandas as pd
```

```python
df_raw = pd.read_csv('data/teelt-gegevens-per-bed-2022.csv', header=[1])
```

```python
df_raw.head()
```

```python
df_raw.columns
```

```python
df1 = df_raw[['soort', 'varieteit', 'voorzaai datum',
       'plant datum', 'direct gezaaid?', 'hoeveel meter?', 'plant aantal',
       'oogst maand', 'productie', 'hoeveel mee genomen?', 'bron',
       'waarom mislukt?', 'opmerkingen']]
df2 = df_raw[['soort.1', 'varieteit.1',
       'voorzaai datum.1', 'plant datum.1', 'direct gezaaid?.1',
       'hoeveel meter?.1', 'plant aantal.1', 'oogst maand.1', 'productie.1',
       'hoeveel meegenomen?', 'bron.1', 'waarom mislukt?.1', 'opmerkingen.1']]
df3 = df_raw[['soort.2', 'varieteit.2', 'voorzaai datum.2', 'plant datum.2',
       'direct gezaaid?.2', 'hoeveel meter?.2', 'plant aantal.2',
       'oogst maand.2', 'productie.2', 'hoeveel meegenomen?.1', 'bron.2',
       'waarom mislukt?.2', 'opmerkingen.2']]
df2.columns = df1.columns
df3.columns = df1.columns

df = pd.concat([df1, df2, df3])
```

```python
df = df.dropna(subset=['soort', 'hoeveel meter?', 'opmerkingen'], how="all")
```

```python
df['soort'] = df['soort'].str.lower()
```

```python
pd.set_option('display.max_rows', 500)
df_sorted = df.sort_values('soort')
df_sorted
```

```python
df_sorted.to_csv('data/teeltgegevens-per-gewas-2022.csv', index=False)
```
